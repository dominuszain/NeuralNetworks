```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras import layers, models
```

```python
training_data = pd.read_csv('neural.csv', names=['Mass','Energy','Residual'])
training_data.head()
```

```python
nuclear_features = training_data.copy()
nuclear_labels = nuclear_features.pop('Residual')
```

```python
nuclear_features = np.array(nuclear_features)
nuclear_labels = np.array(nuclear_labels)
```

```python
nuclear_model = tf.keras.Sequential([
    tf.keras.Input(shape=(2,)),
    layers.Dense(8, activation='relu'),
    layers.Dense(8, activation='relu'),
    layers.Dense(1)
])

nuclear_model.compile(
    loss=tf.keras.losses.MeanSquaredError(),
    optimizer=tf.keras.optimizers.Adam()
)
```

```python
nuclear_model.fit(nuclear_features, nuclear_labels, epochs=1000)
```
